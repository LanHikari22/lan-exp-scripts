```python
import os, sys
parent_dir = os.path.abspath('.')
if parent_dir not in sys.path:
    sys.path.append(parent_dir)
from comm000 import plot
import comm000 as comm

from typing import List, Dict, Tuple
from collections import defaultdict

salaries_and_tenures: List[Tuple[int, float]] = [
    (83000, 8.7), (88000, 8.1),
    (48000, 0.7), (76000, 6),
    (69000, 6.5), (76000, 7.5),(60000, 2.5), (83000, 10),
    (48000, 1.9), (63000, 4.2)
]

salaries_by_tenure: Dict[float, List[int]] = comm.group_list_of_tup2_by_col_with_append(salaries_and_tenures, 1)

salaries_by_tenure
```

```python
xy_obj = plot.XYFloatColumnLists.from_list_of_tup(salaries_and_tenures, reverse=True)

xy_obj.scatterplot_and_display(plot.PlotLineSettings(
    plot.PlotLinePattern.SCATTER, 
    'tenure (yr)',
    'salary ($)',
    ''
))
```

```python
def avg_salary_by_tenure(salaries_by_tenure: Dict[float, List[int]]) -> Dict[float, float]:
    return {
        tenure: sum(salaries) / len(salaries)
            for tenure, salaries in salaries_by_tenure.items()
    }

avg_salary_by_tenure(salaries_by_tenure)
```
