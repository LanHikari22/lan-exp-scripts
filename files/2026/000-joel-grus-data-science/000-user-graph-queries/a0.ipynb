```python
from typing import List, Dict, Tuple
from collections import Counter
from dataclasses import dataclass

type Id = int

@dataclass
class UserRecord:
    id: Id
    name: str

users: List[UserRecord] = [
    UserRecord(id=0, name="Hero"),
    UserRecord(id=1, name="Dunn"),
    UserRecord(id=2, name="Sue"),
    UserRecord(id=3, name="Chi"),
    UserRecord(id=4, name="Thor"),
    UserRecord(id=5, name="Clive"),
    UserRecord(id=6, name="Hicks"),
    UserRecord(id=7, name="Devin"),
    UserRecord(id=8, name="Kate"),
    UserRecord(id=9, name="Klein")
]

friendship_pairs = [(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4),
                    (4, 5), (5, 6), (5, 7), (6, 8), (7, 8), (8, 9)]

friendships: Dict[Id, List[Id]] = {user.id: [] for user in users }

for (i, j) in friendship_pairs:
    friendships[i].append(j)
    friendships[j].append(i)

interests: List[Tuple[Id, str]] = [
    (0, "Hadoop"), (0, "Big Data"), (0, "HBase"), (0, "Java"),
    (0, "Spark"), (0, "Storm"), (0, "Cassandra"),
    (1, "NoSQL"), (1, "MongoDB"), (1, "Cassandra"), (1, "HBase"),
    (1, "Postgres"), (2, "Python"), (2, "scikit-learn"), (2, "scipy"),
    (2, "numpy"), (2, "statsmodels"), (2, "pandas"), (3, "R"), (3, "Python"),
    (3, "statistics"), (3, "regression"), (3, "probability"),
    (4, "machine learning"), (4, "regression"), (4, "decision trees"),
    (4, "libsvm"), (5, "Python"), (5, "R"), (5, "Java"), (5, "C++"),
    (5, "Haskell"), (5, "programming languages"), (6, "statistics"),
    (6, "probability"), (6, "mathematics"), (6, "theory"),
    (7, "machine learning"), (7, "scikit-learn"), (7, "Mahout"),
    (7, "neural networks"), (8, "neural networks"), (8, "deep learning"),
    (8, "Big Data"), (8, "artificial intelligence"), (9, "Hadoop"),
    (9, "Java"), (9, "MapReduce"), (9, "Big Data")
]
```

```python
def num_friends(user: UserRecord) -> int:
    return len(friendships[user.id])

def _total_num_of_connections() -> int:
    return sum([len(friendships[user_id]) for user_id in friendships])

def avg_num_of_connections() -> float:
    return _total_num_of_connections() / len(users)

def _get_user_and_friend_count_tup() -> List[Tuple[Id, int]]:
    return [(user_id, len(friendships[user_id])) for user_id in friendships]

def sorted_by_popularity() -> List[Tuple[Id, int]]:
    return sorted(_get_user_and_friend_count_tup(), key=lambda user_num_friends: user_num_friends[1], reverse=True)

def mutual_friend_count(user: UserRecord) -> Counter[Id]:
    # foaf is friend of a friend
    return Counter([
        foaf_id
            for friend_id in friendships[user.id]
            for foaf_id in friendships[friend_id]
            if foaf_id != user.id and 
               foaf_id not in friendships[user.id]
    ])

def users_who_like(interest: str) -> List[Id]:
    return [
        user_id
            for (user_id, cur_interest) in interests
            if cur_interest == interest
    ]

users_who_like("Big Data")
```
